# IRIS Data Classification

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset:
from sklearn.datasets import load_iris
```

```python
#load dataset
data=load_iris()
data
```

```python
#converting data into dataframe:
data_x=pd.DataFrame(data=data.data,columns=data.feature_names)
data_x
```

```python
#all information about data:
data_x.info()
```

```python
#describe datasets:
data_x.describe()
```

```python
#addong target variables into datafrme:
data_y=pd.DataFrame(data=data.target,columns=["class"])
data_y
```

```python
#renaming the names of column:
data_x.rename(columns = {
    'sepal length (cm)' : 'sepal_length',
    'sepal width (cm)' : 'sepal_width',
    'petal length (cm)' : 'petal_length',
    'petal width (cm)' : 'petal_width'
},inplace = True)
data_x
```

```python
#split data 
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(data_x, data_y, test_size=0.2)
```

```python
#prepare model using LogisticRegression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
#predict the x test:
pred=lr.predict(xtest)
pred
```

```python
#converting predicted data into into dataframe
pred= pd.DataFrame(pred)
pred.head()
```

```python
ytest.head()
```

```python
#importing accuracy_score, classification_report, confusion_matrix from sklearn
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#for accuracy_score:
ac = accuracy_score(ytest, pred)
print("accuracy score is :",ac)
```

```python
#for classification_report:
cr=classification_report(ytest,pred)
print("classification_report",cr)
```

```python
#for confusion_matrix:
cm=confusion_matrix(ytest,pred)
print(cm)
```
